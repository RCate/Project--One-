```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
in_file = open("test.htm", "r")
in_file.readline()
in_file.readline()
my_str = in_file.readline()
```

```python
my_list = my_str.split('"comments":')
```

```python
my_list = my_list[1:]
```

```python
comments = [s.split('"id":')[0] for s in my_list]
```

```python
for e in comments:
    print(e)
```

```python
comments
```

```python
wordcloud = WordCloud().generate(comments[5])
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(comments[4])
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
