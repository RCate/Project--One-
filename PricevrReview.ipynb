```python
import gmaps
import gmaps.datasets
import numpy as np
import requests
import json
import pandas as pd
import base64
import matplotlib.pyplot as plt
```

```python
# Reading in the csv file
air = pd.read_csv("Airbnb_Austin_Data_Finalized.csv")
air.head()
```

```python
# np.where(pd.isnull(airbig))
air.dropna(subset=['price'], inplace=True)
air.dropna(subset=['review_scores_rating'], inplace=True)
air.head(1)
```

```python
# Convert to a dataframe.
df = pd.DataFrame(air)
```

```python
# defining variables to the scores and price columns in df
rating = df["review_scores_rating"]
price = df["price"]
```

```python
# creating a scatter plot
plt.scatter(x=rating, y=price, c="red", edgecolors='k')


# Set a Title, axis labels, legend
plt.title("Price vrs Reviews Score")
plt.xlabel("Review Value")
plt.ylabel("Price ($)")

# pylab.legend(loc=9, bbox_to_anchor=(1.15, 1))
plt.savefig('Plot_PricevrReviews.png')
```
